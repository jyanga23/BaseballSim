```julia
function at_bat(curr_state)
    
    outcomes = transition_matrix[get_state_index(curr_state)]
    prob_space = cumsum(vec(outcomes))
    # ??? next_state_ind comes back undefined when curr_state == 3, 4, 11, 12
    next_state_ind = findfirst(rand() .< prob_space)
    next_state = state_dict[next_state_ind][2]
    
    return next_state
    
end;



#=
for state in state_dict
    println(state[1])
    test_state = state[2]
    outcomes = transition_matrix[get_state_index(test_state)]
    println("outcomes: ", outcomes)
    prob_space = cumsum(vec(outcomes))
    println("ps: ", prob_space)
    next_state_ind = findfirst(rand() .< prob_space)
    println("next ind: ", next_state_ind, "\n")
end
=#
```

```julia
function check_score(new_state, old_state, pa, runs)
    outs = new_state[1] - old_state[1]
    pa -= outs
    players_on_base = count(new_state[2])
    new_runs = runs + (pa - players_on_base)
    if (new_runs != runs) pa -= (pa - players_on_base) end
    
    return new_runs, pa
end;
```

```julia
function half_inning(score)
    curr_state = (0, (false, false, false))
    runs = 0
    players_active = 0
    players_on_base = 0
    
    while curr_state[1] < 3
        players_active += 1
        new_state = at_bat(curr_state)
        runs, players_active = check_score(new_state, curr_state, players_active, runs)
        curr_state = new_state
    end

    return runs
    
end;
```

```julia
function game_simulation()
    home_score = 0
    away_score = 0
    
    num_innings = 9
    
    for i in 1:num_innings
        println("Inning ", i, "\n")
        home_score += half_inning(home_score)
        println("Home Score: ", home_score)
        away_score += half_inning(away_score)
        println("Away Score: ", away_score)
        println("\n-----------------------------------\n\n\n")
    end
    
    return home_score, away_score
end
```
