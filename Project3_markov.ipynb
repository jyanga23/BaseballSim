```julia
function at_bat(curr_state, team, batter)    
    outcomes = team[batter][get_state_index(curr_state)]
    prob_space = cumsum(vec(outcomes))
    next_state_ind = findfirst(rand() .< prob_space)
    next_state = state_dict[next_state_ind][2]
    
    return next_state
    
end;
```

```julia
function check_score(new_state, old_state, pa, runs)
    outs = new_state[1] - old_state[1]
    pa -= outs
    players_on_base = count(new_state[2])
    new_runs = runs + (pa - players_on_base)
    if (new_runs != runs) pa -= (pa - players_on_base) end
    
    return new_runs, pa
end;
```

```julia
function half_inning(score, team, batter)
    curr_state = (0, (false, false, false))
    runs = 0
    players_active = 0
    players_on_base = 0
    
    while curr_state[1] < 3
        players_active += 1
        new_state = at_bat(curr_state, team, batter)
        runs, players_active = check_score(new_state, curr_state, players_active, runs)
        curr_state = new_state
    end

    return runs
    
end;
```

```julia
function game_simulation(home_team, away_team)
    home_score = 0
    home_batter = 0
    
    away_score = 0
    away_batter = 0
    
    num_innings = 9
    
    for i in 1:num_innings
        #println("Inning ", i, "\n")
        
        # Batting rotation continues from where it left off? Batting rotation starts over?
        home_batter += 1
        if (home_batter > 9) home_batter = 1 end
        home_score += half_inning(home_score, home_team, home_batter)
        #println("Home Score: ", home_score)
        
        away_batter += 1
        if (away_batter > 9) away_batter = 1 end
        away_score += half_inning(away_score, away_team, away_batter)
        #println("Away Score: ", away_score)
        
        #println("\n-----------------------------------\n\n\n")
    end
    
    return home_score, away_score
end
```

```julia

function simulate_season()
    results = Dict()
    wins = Dict(team[1] => 0 for team in teams)

    for i in 1:length(teams)
        num_games = 100

        total_home_score = 0
        total_away_score = 0
        for game in 1:num_games
            home_score, away_score = game_simulation(home_team, teams[i][2])
            total_home_score += home_score
            total_away_score += away_score

            if (away_score > home_score)
                wins[teams[i][1]] += 1
            end
        end

        avg_home_score = total_home_score / num_games
        avg_away_score = total_away_score / num_games

        results[teams[i][1]] = avg_away_score
    end
    
    return results, wins
    
end
```
